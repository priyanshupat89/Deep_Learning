split()

```python
cv2_imshow(img_color)
print(type(img_color))
print(img_color)
b = img_color[:,:,0]
g = img_color[:,:,1]
r = img_color[:,:,2]
print(img_color.shape, b.shape, g.shape, r.shape)
cv2_imshow(b)
cv2_imshow(g)
cv2_imshow(r)
```

```python
cv2_imshow(img_color)
print(type(img_color))
print(img_color)
b,g,r = cv2.split(img_color)
print(img_color.shape, b.shape, g.shape, r.shape)
cv2_imshow(b)
cv2_imshow(g)
cv2_imshow(r)
```

```python
merged_img = cv2.merge((b,g,r))
cv2_imshow(merged_img)
```
