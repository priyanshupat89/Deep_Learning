mini batch


```python
class ANN:
    def __init__(self):
        self.w1 = 1
        self.w2 = 1
        self.b = 0

    def fit(self, x, y, epochs, batch_size=32, lr=0.5):
        self.mini_batch_gradient_descent(x['age'], x['affordibility'], y, epochs, batch_size, lr)

    def mini_batch_gradient_descent(self, age, aff, y_t, epochs, batch_size, lr):
        w1 = 1
        w2 = 1
        b = 0
        n = len(age)
        age = np.array(age)
        y_t = np.array(y_t)
        aff = np.array(aff)

        for epoch in range(epochs):
            indices = np.arange(n)
            np.random.shuffle(indices)

            for start in range(0, n, batch_size):
                end = min(start + batch_size, n)
                batch_indices = indices[start:end]

                batch_age = age[batch_indices]
                batch_aff = aff[batch_indices]
                batch_y_t = y_t[batch_indices]

                W_S = w1 * batch_age + w2 * batch_aff + b
                y_p = sigmoid(W_S)
                loss = log_loss(batch_y_t, y_p)

                w_age = np.sum(batch_age * (y_p - batch_y_t))
                w_aff = np.sum(batch_aff * (y_p - batch_y_t))
                bais = np.sum(batch_y_t - y_p)

                w1 = w1 - lr * w_age / len(batch_indices)
                w2 = w2 - lr * w_aff / len(batch_indices)
                b = b - lr * bais / len(batch_indices)

            print("Epochs-", epoch, " ", "W1-", w1, " ", "w2-", w2, " ", "bais-", b, "Loss :- ", loss)

        self.w1, self.w2, self.b = w1, w2, b

    def predict(self, x):
        vlu = self.w1 * x['age'] + self.w2 * x['affordibility'] + self.b
        y_p = np.array(sigmoid(vlu))
        return y_p
```
